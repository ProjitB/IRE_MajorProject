```python
import os
import random
import nltk
from nltk.corpus import reuters
from nltk import bigrams, trigrams
from collections import Counter, defaultdict
```

#### Dataset Link

##### http://www.statmt.org/wmt11/translation-task.html#download
Downloaded the 2010 version which had 2 GB of english news text. (after extraction)

```python
def read_data(data_path):
    with open(data_path + "/" + "news.2010.en.shuffled") as news_data_file:
        c = 0
        data = []
        while c < 300000:
            data.append(news_data_file.readline())
            c += 1
    return data
```

```python
def get_bigrams(data):
    actual_data = [line.split() for line in data if line is not None]
    bi_grams = []
    for line in actual_data:
        bi_grams += list(bigrams(line))
    return bi_grams
```

```python
def make_language_model(bi_grams):
    model = defaultdict(lambda: defaultdict(lambda: 0))
    for w1, w2 in bi_grams:
        model[w1][w2] += 1
        
    for w1 in model:
        total = float(sum(model[w1].values()))
        for w2 in model[w1]:
            model[w1][w2] /= total
    return model
```

```python
def show_model(model):
    for w1,w2_map in model.items():
        for w2,prob in w2_map.items():
            print(w1, w2, prob)
```

#### Main Program

```python
data_path = "/media/kaushik/Studies/IIITH/3_ThirdSem/IRE/Major Project/training-monolingual"
data = read_data(data_path)
```

```python
bi_grams = get_bigrams(data)
```

```python
model = make_language_model(bi_grams)
```

```python
show_model(model)
```
